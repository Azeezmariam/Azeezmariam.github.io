```python
# YOUR CODE HERE
# Plot a histogram

# Set the figure size
plt.figure(figsize=(10, 6))

# Plot the histogram with different bin widths
plt.hist(data['co2'], bins=20, color='yellow', alpha=0.9, label='Bin width = 25')
plt.hist(data['co2'], bins=50, color='grey', alpha=0.7, label='Bin width = 50')
plt.hist(data['co2'], bins=100, color='green', alpha=0.3, label='Bin width = 75')

# Add labels and title
plt.title('Distribution of carbon Dioxide ')
plt.xlabel('Carbon Dioxide  (g/mi)')
plt.ylabel('Frequency')


```

```python
# YOUR CODE HERE
# Select only numeric columns
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr_matrix = numeric_data.corr()

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
